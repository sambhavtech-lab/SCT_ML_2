Python Code for K-Means

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Ensure 'data' is the original DataFrame
data = pd.read_csv("/content/drive/MyDrive/ML_Projects/customer.csv")

# Prepare data for clustering (similar to AKlK8OqkdsR9)
X = data.drop("CustomerID", axis=1)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply K-means clustering (similar to FIkTIPPre2Cj)
kmeans = KMeans(n_clusters=3, random_state=42, n_init='auto') # Added n_init='auto' to suppress warning
data["Cluster"] = kmeans.fit_predict(X_scaled)

print(data.columns)
plt.scatter(data['AnnualSpend '], data['visit frequency'], c=data['Cluster'])
plt.xlabel('AnnualSpend')
plt.ylabel('visit frequency')
plt.title('Customer Segmentation (kmeans)')
plt.show()
```

Elbow Method

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import pandas as pd

# Create dataframe
data = pd.DataFrame({
    'AnnualSpend':[20000,50000,15000,80000,30000,90000,25000,60000],
    'VisitFrequency':[12,30,10,40,20,45,15,35],
    'AvgBasketValue':[1500,2500,1200,3000,1800,3200,1600,2700]
})

# Scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(data)

# Elbow Method
wcss = []

# Adjust the range to not exceed the number of samples
for i in range(1, X_scaled.shape[0] + 1): # Max clusters should be less than or equal to n_samples
    km = KMeans(n_clusters=i, random_state=42, n_init='auto') # Added n_init='auto'
    km.fit(X_scaled)
    wcss.append(km.inertia_)

plt.plot(range(1, X_scaled.shape[0] + 1), wcss, marker='o')
plt.xlabel("Clusters")
plt.ylabel("WCSS")
plt.title("Elbow Method")
plt.show()
```
